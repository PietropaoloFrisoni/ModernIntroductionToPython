### The global namespace

```python
x = 42


def my_function():
    pass


class MyClass:
    pass
```

```python
global_names = list(globals().keys())
```

```python
print(f"x in global_names: {"x" in global_names}")
print(f"my_function in global_names: {"my_function" in global_names}")
print(f"MyClass in global_names: {"MyClass" in global_names}")
```

### The local namespace

```python
def show_locals():
    a = "local a"
    b = "local b"
    c = "local c"
    print(f"Local namespace: {locals()}")
```

```python
show_locals()
```

```python
class MyClass:
    x = 10
    y = 20


obj = MyClass()
obj.z = 30
```

```python
print(f"Class namespace: {MyClass.__dict__}")
print(f"Instance namespace: {obj.__dict__}")
```

## Scopes

```python
# B - Built-in
print(len([1, 2, 3]))  # len is in the built-in namespace

# G - Global
x = "global x"


def outer():
    # E - Enclosing
    x = "enclosing x"

    def inner():
        # L - Local
        x = "local x"
        print(f"inner sees: {x}")

    inner()
    print(f"outer sees: {x}")


outer()
print(f"module sees: {x}")
```

## `global` and `nonlocal`

```python
counter = 0


def increment():
    global counter
    counter += 1
```

```python
increment()
increment()
increment()
```

```python
print(counter)
```

```python
def make_counter():

    count = 0

    def increment():
        nonlocal count
        count += 1
        return count

    return increment
```

```python
counter = make_counter()
```

```python
print(counter())
print(counter())
print(counter())
```

```python
locals() is globals()
```
