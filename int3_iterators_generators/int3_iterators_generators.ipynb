## Generator Functions

```python
def count_up(n):
    print("Starting generator")
    for i in range(n):
        print(f"   About to yield {i}")
        yield i
        print(f"   Resumed after yielding {i}")
    print("Generator exhausted")
```

```python
type(count_up)
```

```python
gen = count_up(3)
print(type(gen))
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
# print(next(gen))   this raises a StopIteration error
```

```python
for val in count_up(3):
    print(val)
```

## Generator Expressions

```python
squares_list = [x**2 for x in range(10)]
print(f"List: {squares_list}")

squares_gen = (x**2 for x in range(10))
print(f"Generator: {squares_gen}")
```

```python
for square in squares_gen:
    print(square)
```

```python
import sys

N = 1_000_000
print(f"List: {sys.getsizeof([x**2 for x in range(N)])} bytes")
print(f"Generator: {sys.getsizeof((x**2 for x in range(N)))} bytes")
```

## Practical Uses of Generators

```python
def count_forever(start=0):
    n = start
    while True:
        yield n
        n += 1
```

```python
counter = count_forever()
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
print(next(counter))
# ...
```

## The `__iter__` and `__next__` Methods (Preview)

```python
class Counter:

    def __init__(self, start, stop):
        self.current = start
        self.stop = stop

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.stop:
            raise StopIteration
        value = self.current
        self.current += 1
        return value
```

```python
for n in Counter(1, 5):
    print(n)
```
