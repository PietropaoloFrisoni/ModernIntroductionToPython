# Chapter 1: Python as a Calculator


## Numbers and Arithmetic

```python
print(f"Addition: {3+5}")
print(f"Subtraction: {8-5}")
print(f"Multiplication: {-8 * 4}")
print(f"Division: {4/2}")
```

## Floor Division And Remainder

```python
print(f"Division (7/2): {7/2}")
print(f"Floor division (7//2): {7//2}")
print(f"Remainder (7%2): {7%2}")
```

## Exponentiation

```python
print(f"2 to the power of 3: {2 ** 3}")
print(f"Square root of 16: {16 ** 0.5}")
print(f"2 to the power of 0: {2 ** 0}")
```

## Variables and Operator Precedence


```python
a = 3 + 4 * 2
b = (3 + 4) * 2
print(f"Without parenthesis: {a}")
print(f"With parenthesis: {b}")
```

## Comparison and Logical Operators

### Comparison operators

```python
a = 3
b = 5

print(f"a == b: {a == b}")  # equality
print(f"a != b: {a != b}")  # inequality
print(f"a < b: {a < b}")  # less than
print(f"a > b: {a > b}")  # greater than
print(f"a <= b: {a <= b}")  # less than or equal
print(f"a >= b: {a >= b}")  # greater than or equal
```

### Logical operators

```python
x = 4

print(x > 0 and x < 10)  # True if both are True
print(x < 0 or x == 4)  # True if at least one is True
print(not (x == 4))  # reverses the boolean value
```

## Special Values: `None`, `NaN`, and `Infinity`

### None

```python
x = None
print(f"Value of x: {x}")
print(f"Type of x: {type(x)}")
```

```python
print(x is None)
print(x is not None)
```

### NaN

```python
import math

nan = math.nan
print(f"Value of nan: {nan}")
print(f"Type of nan: {type(nan)}")
```

```python
print(nan is nan)
print(nan == nan)
```

```python
print(math.isnan(nan))
```

```python
print(nan + 1)
print(nan * 0)
print(nan > 0)
```

### Infinity

```python
pos_inf = math.inf
neg_inf = -math.inf
```

```python
print(f"Positive infinity: {pos_inf}")
print(f"Negative infinity: {neg_inf}")
print(f"Type of positive infinity: {type(pos_inf)}")
```

```python
print(pos_inf + 1000)
print(pos_inf > 1000)
print(pos_inf - pos_inf)
print(pos_inf * 0)
```

```python
print(math.isinf(pos_inf))
print(math.isinf(42))
```

### Final Summary

```python
print(f"None - absence of a value: {None}")
print(f"NaN - undefined numeric result: {math.nan}")
print(f"Inf - unbounded numeric value: {math.inf}")
```

## Useful Built-ins: `round` and `abs`

### `round`

```python
pi = 3.141592

print(f"round(pi): {round(pi)}")
print(f"round(pi, 2): {round(pi, 2)}")
print(f"round(pi, 4): {round(pi, 4)}")
```

### `abs`

```python
print(f"abs(-5): {abs(-5)}")
print(f"abs(-5.3456): {abs(-5.3456)}")
print(f"abs(3+4j): {abs(3+4j)}")
```
