# Intermezzo 1: Booleans in Depth

## Truthiness

### Falsy values

```python
print(bool(0))
print(bool(0.0))
print(bool(""))
print(bool([]))
print(bool({}))
print(bool(set()))
print(bool(None))
```

### Truthy values

```python
print(bool(1))
print(bool(-1))
print(bool("hello"))
print(bool([0]))
```

```python
items = [1, 2, 3, 4, 5]

if len(items) > 0:
    pass

# equivalent to the above (more Pythonic)
if items:
    pass
```

## Short-Circuit Evaluation

```python
# equivalent to print(False)
print(False and print("This is not going to be displayed"))
```

```python
# equivalent to print(True)
print(True or print("This is not going to be displayed"))
```

```python
data = None

if data is not None and data["key"] > 0:
    print("This is not going to be displayed")
```

## `any` and `all`

### `any`

```python
values = [3, 7, 2, 9, 1]

found = False
for v in values:
    if v > 5:
        found = True
        break
print(found)
```

```python
# equivalent version with `any`
print(any(v > 5 for v in values))
```

### `all`

```python
print(all(v > 0 for v in values))
print(all(v > 5 for v in values))
```

Edge cases

```python
print(any([]))
print(all([]))
```

## Generator Expressions (first look)

```python
squares_list = [x**2 for x in range(10)]
print(squares_list)
```

```python
squares_gen = (x**2 for x in range(10))
print(squares_gen)
```

```python
for sq in squares_gen:
    print(sq)
```
