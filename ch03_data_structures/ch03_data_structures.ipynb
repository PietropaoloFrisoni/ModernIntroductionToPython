## `deque` from collections

```python
from collections import deque

dq = deque([1, 2, 3])
print(f"Deque: {dq}")

dq.append(4)
print(f"Deque after append: {dq}")
dq.appendleft(0)
print(f"Deque after appendleft: {dq}")

dq.pop()
print(f"Deque after pop: {dq}")
dq.popleft()
print(f"Deque after popleft: {dq}")
```

```python
import time

N = 100_000

# List: insert at beginning — O(n) each time
start = time.perf_counter()
lst = []
for n in range(N):
    lst.insert(0, n)
end = time.perf_counter()
print(f"List insert: {end - start:.4f} seconds")

# Deque: appendleft — O(1) each time
start = time.perf_counter()
dq = deque()
for n in range(N):
    dq.appendleft(n)
end = time.perf_counter()
print(f"Deque appendleft: {end - start:.4f} seconds")
```

### namedtuple (bonus)

```python
from collections import namedtuple

Point = namedtuple("Point", ["x", "y", "z"])
p = Point(10, 20, 30)

print(f"Point: {p}")
print(f"x: {p.x}")
print(f"y: {p.y}")
print(f"z: {p[2]}")
```

## Dictionaries

```python
person = {"name": "Pietro", "age": 31, "city": "Munich"}
print(f"Dictionary: {person}")
```

```python
person["profession"] = "Quantum Software Engineer"
print(f"Dictionary: {person}")
```

```python
print(f"Name: {person["name"]}")
print(f"Age: {person["age"]}")
```

```python
print(f"Name: {person.get("name")}")
print(f"Age: {person.get("age")}")
print(f"Salary: {person.get("salary", "N/A")}")
```

```python
for key in person:
    print(f"key: {key}")

print("----")

for key in person.keys():
    print(f"key: {key}")
```

```python
for value in person.values():
    print(f"value: {value}")
```

```python
for key, value in person.items():
    print(f"key: {key}, value: {value}")
```

```python
squares = {x: x**2 for x in range(6)}
print(f"Squares: {squares}")
```

```python
d = {}
d[(1, 2)] = "this is going to work"
# d[[1,2]] = "this is not going to work"
print(d)
```
