## Basic Inheritance

```python
class Animal:

    species = "Animal"

    def __init__(self, name, sound) -> None:
        print("__init__ in Animal has been called")
        print(f"self: {self}")
        self.name = name
        self.sound = sound

    def speak(self):
        print(f"{self.name} says: {self.sound}")

    def describe(self):
        print(f"I am {self.name}, an {self.species}")
```

```python
class Dog(Animal):

    def fetch(self):
        print(f"self: {self}")
        print(f"{self.name} fetches the ball!")
```

```python
dog = Dog(name="Rex", sound="Woof")
```

```python
dog.fetch()
```

```python
# dir(dog)
```

```python
dog.speak()
dog.describe()
print(dog.species)
```

```python
print(isinstance(dog, Dog))
print(isinstance(dog, Animal))
```

## Method Overriding and `super()`

```python
class Animal:

    def __init__(self, name) -> None:
        print("__init__ in Animal has been called")
        print(f"self: {self}")
        self.name = name

    def speak(self):
        print(f"self: {self}")
        print(f"{self.name} makes a sound")
```

```python
class Cat(Animal):

    def speak(self):
        super().speak()
        print(f"{self.name} says: Meow!")
        print(f"self: {self}")
```

```python
cat = Cat("Whiskers")
```

```python
cat.speak()
```

```python
class Dog(Animal):

    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def speak(self):
        super().speak()
        print(f"{self.name} is a {self.breed}")
```

```python
dog = Dog(name="Rex", breed="Labrador")
```

```python
dog.speak()
print(dog.name)
print(dog.breed)
```

## Multiple Inheritance

```python
class Animal:

    def __init__(self, name, sound):
        print("__init__ in Animal has been called")
        print(f"self: {self}")
        self.name = name
        self.sound = sound

    def speak(self):
        print(f"{self.name} says: {self.sound}")
```

```python
class Flyable:

    def fly(self):
        print(f"self: {self}")
        # Typo in recording: "if" should be "is"
        print(f"{self.name} is flying!")

    def generic_method(self):
        print("generic method called in the Flyable class")
```

## The Diamond Problem

```python
class Grandparent:

    def greet(self):
        print("Hello from Grandparent!")
        print(f"self: {self}")


class Parent1(Grandparent):

    def greet(self):
        print("Hello from Parent1!")
        super().greet()


class Parent2(Grandparent):

    def greet(self):
        print("Hello from Parent2!")
        super().greet()


class Child(Parent1, Parent2):

    def greet(self):
        print("Hello from Child!")
        super().greet()
```

```python
Child().greet()
```

```python
Grandparent().greet()
```

```python
Child.mro()
```

## Mixins

```python
import datetime


class LoggingMixin:

    def log(self, message):
        print(f"[{type(self).__name__}] {message}")


class TimestampMixin:

    def timestamp(self):
        return datetime.datetime.now().isoformat()


class ValidationMixin:

    def validate(self, value, minimum, maximum):
        return minimum <= value <= maximum
```

```python
class Sensor(LoggingMixin, TimestampMixin, ValidationMixin):

    def __init__(self, name, min_val, max_val):
        self.name = name
        self.min_val = min_val
        self.max_val = max_val

    def read(self, value):
        if self.validate(value, self.min_val, self.max_val):
            self.log(f"Reading at {self.timestamp()}: {value}")
        else:
            self.log(f"Value {value} out of range!")
```

```python
sensor = Sensor("Temperature", -10, 100)
sensor.read(25.5)
```

```python
sensor.read(200)
```

## Abstract Classes

```python
from abc import ABC, abstractmethod


class Shape(ABC):

    @property
    @abstractmethod
    def name(self) -> str:
        pass

    @abstractmethod
    def area(self) -> float:
        """Return the area of the shape."""
        pass

    @abstractmethod
    def perimeter(self) -> float:
        """Return the perimeter of the shape."""
        pass

    def describe(self) -> str:
        """Describe the shape."""
        return f"Shape with area={self.area():.2f} and perimeter={self.perimeter():.2f}"
```

```python
import math


class Circle(Shape):

    def __init__(self, radius: float) -> None:
        self.radius = radius

    @property
    def name(self) -> str:
        return "Circle"

    def area(self) -> float:
        return math.pi * self.radius**2

    def perimeter(self) -> float:
        return 2 * math.pi * self.radius
```

```python
class Rectangle(Shape):

    def __init__(self, width: float, height: float) -> None:
        self.width = width
        self.height = height

    @property
    def name(self) -> str:
        return "Rectangle"

    def area(self) -> float:
        return self.width * self.height

    def perimeter(self) -> float:
        return 2 * (self.width + self.height)
```

```python
c = Circle(5)
print(c.describe())
print(c.name)
```

```python
r = Rectangle(4, 6)
print(r.describe())
print(r.name)
```

## Polymorphism and Duck Typing

```python
class Circle:

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2
```

```python
class Rectangle:

    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height
```

```python
class Triangle:

    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height
```

```python
def print_area(shape):
    print(f"{type(shape).__name__}: area = {shape.area():.2f}")
```

```python
shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 8)]
for shape in shapes:
    print_area(shape)
```
