## The Argument Passing Model

```python
outer_x = 5


def fun(inner_x):
    print(f"Before: inner_x is outer_x? {inner_x is outer_x}")
    inner_x += 1
    print(f"After: inner_x is outer_x? {inner_x is outer_x}")
```

```python
print(f"outer_x before: {outer_x}")
fun(outer_x)
print(f"outer_x after: {outer_x}")
```

```python
outer_x = [5]


def fun(inner_x):
    print(f"Before: inner_x is outer_x? {inner_x is outer_x}")
    inner_x.append(7)
    print(f"After: inner_x is outer_x? {inner_x is outer_x}")
```

```python
print(f"outer_x before: {outer_x}")
fun(outer_x)
print(f"outer_x after: {outer_x}")
```

## Returning Values

```python
def square(x):
    return x * x
```

```python
result = square(4)
print(f"Result: {result}")
```

```python
def square_no_return(x):
    print(x * x)
    return None
```

```python
result = square_no_return(4)
print(f"Result: {result}")
```

```python
def min_max(lst):
    return min(lst), max(lst)
```

```python
min, max = min_max([1, 2, 3])
print(f"Min: {min}, Max: {max}")
```

## Why Functions?

```python
def celsius(fahrenheit):
    return (fahrenheit - 32) * 5 / 9
```

```python
temps_f = [32, 68, 45, 78]
temps_c = [celsius(t) for t in temps_f]
print(f"Temperatures in celsius: {temps_c}")
```

## The Mutable Default Argument Trap

```python
def add_item(item, collection=[]):
    collection.append(item)
    return collection
```

```python
print(add_item("apple"))
print(add_item("banana"))
print(add_item("cherry"))
```

```python
def add_item(item, collection=None):
    collection = [] if collection is None else collection
    collection.append(item)
    return collection
```

```python
print(add_item("apple"))
print(add_item("banana"))
print(add_item("cherry"))
```

## Keyword Arguments

```python
def make_sandwich(bread, filling, toasted=False):
    print(f"Making a {filling} sandwich on {bread} bread.")
    if toasted:
        print("...toasted")
```

```python
make_sandwich("rye", "cheese")
```

```python
make_sandwich("rye", "cheese", toasted=True)
```

```python
make_sandwich(filling="cheese", toasted=True, bread="rye")
```

## Packing and Unpacking (*args and **kwargs)

```python
def pack(*args, **kwargs):
    print(f"args: {args} (type: {type(args)})")
    print(f"kwargs: {kwargs} (type: {type(kwargs)})")
```

```python
pack(1, 2, 3, 4, user="Richard", operation="Sum")
```

```python
def greet(name, message):
    print(f"{message}, {name}!")
```

```python
args = ("Alice", "Welcome")
kwargs = {"name": "Bob", "message": "Good morning"}
```

```python
greet(*args)  # equivalent to greet("Alice", "Welcome")
greet(**kwargs)  # equivalent to greet(name="Bob", message="Good morning")
```

## Docstrings

```python
from math import pi


def area_of_circle(radius):
    """
    Computes the area of a circle given its radius.

    Parameters
    ----------
    radius : float
        The radius of the circle

    Returns
    -------
    float
        The computed area
    """
    return pi * radius**2
```

```python
area_of_circle(5.5)
```

```python
print(area_of_circle.__doc__)
```

```python
import inspect

print(inspect.getdoc(area_of_circle))
```

```python
help(area_of_circle)
```
