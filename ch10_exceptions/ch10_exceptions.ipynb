## `try`/`except`

```python
def safe_divide(a, b):
    try:
        result = a / b
        print(f"{a} / {b} = {result}")
    except ZeroDivisionError:
        print("Error: cannot divide by zero!")
```

```python
safe_divide(10, 2)
```

```python
safe_divide(10, 0)
```

```python
try:
    x = int("string")
except ValueError as e:
    print(f"Caught ValueError: {e}")
    print(f"Exception type: {type(e).__name__}")
```

```python
try:
    print(undefined_variable)
except NameError as e:
    print(f"NameError: {e}")
```

```python
try:
    10 / 0
except ZeroDivisionError as e:
    print(f"ZeroDivisionError: {e}")
```

```python
try:
    "hello" + 5
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
try:
    lst = [1, 2, 3]
    lst[10]
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
try:
    d = {"a": 1}
    print(d["b"])
except KeyError as e:
    print(f"KeyError: {e}")
```

## Catching Multiple Exceptions

```python
def parse_and_divide(num, den):
    try:
        a = int(num)
        b = int(den)
        return a / b
    except ValueError as e:
        print(f"Invalid input: {e}")
    except ZeroDivisionError as e:
        print(f"Cannot divide by zero: {e}")
```

```python
parse_and_divide("1", "0")
```

```python
try:
    result = 2 / 0
except (ValueError, ZeroDivisionError) as e:
    print(f"Something went wrong: {e}")
```

```python
try:
    lst = [1, 2, 3]
    lst[10]
except:
    print("Something happened")
```

## `else` and `finally`

```python
def read_integer(prompt):
    try:
        value = int(input(prompt))
    except ValueError:
        print("This was not a valid integer.")
    else:
        print(f"Successfully read: {value}")
```

```python
read_integer(2)
```

```python
def risky_operation(x):
    print("Starting operation...")
    try:
        result = 100 / x
        print(f"result: {result}")
        return result
    except ZeroDivisionError:
        print("Division by zero!")
        return None
    finally:
        print("Cleanup: this ALWAYS runs.")
```

```python
risky_operation(0)
```

```python
try:
    value = 10 / 0
except ZeroDivisionError:
    print("Division by zero!")
else:
    print(f"value: {value}")
finally:
    print("Cleanup: this ALWAYS runs.")
```

## Raising Exceptions

```python
def set_age(age: int) -> None:
    if not isinstance(age, int):
        raise TypeError(f"age must be an int, got {type(age).__name__}")
    if age < 0:
        raise ValueError(f"age cannot be negative: {age}")
    if age > 150:
        raise ValueError(f"age seems unrealistic: {age}")
    print(f"Age set to {age}")
```

```python
set_age(24)
```

```python
try:
    # set_age("abc")
    set_age(1500)
except ValueError as e:
    print(f"Invalid input: {e}")
    # raise
except TypeError as e:
    print(f"TypeError: {e}")
    # raise
```

## Exception Chaining

```python
def fetch_user(user_id):
    try:
        data = {}
        return data[user_id]
    except KeyError as e:
        raise RuntimeError("User not found") from e
```

```python
try:
    fetch_user("Bob")
except RuntimeError as e:
    print(f"RuntimeError: {e}")
    print(f"Caused by: {e.__cause__}")
```

## Custom Exceptions

```python
class CourseError(Exception):
    """Base exception for course-related errors."""


class EnrollmentError(CourseError):
    """Raised when enrollment fails."""

    def __init__(self, course_name, reason):
        self.course_name = course_name
        self.reason = reason
        super().__init__(f"Cannot enroll in '{course_name}': {reason}")


class CapacityError(EnrollmentError):
    """Raised when a course is full."""

    def __init__(self, course_name, capacity):
        self.capacity = capacity
        super().__init__(course_name, f"course is full ({capacity} students)")
```

```python
def enroll(student, course_name, current_count, max_count):
    if current_count >= max_count:
        raise CapacityError(course_name, max_count)
    print(f"{student} enrolled in {course_name}")
```

```python
try:
    enroll("Alice", "Python", 30, 30)
except CapacityError as e:
    print(f"Capacity error: {e}")
except EnrollmentError as e:
    print(f"Enrollment error: {e}")
```
