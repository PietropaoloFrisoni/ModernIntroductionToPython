## Initializing an Instance

```python
class MyClass:

    def __init__(self, var) -> None:
        print("__init__ has been called")
        print(f"self: {self}")
        print(f"type of self: {type(self)}")
        self.instance_variable = var
```

```python
instance1 = MyClass(5)
print(f"instance1.instance_variable: {instance1.instance_variable}")
```

```python
instance2 = MyClass(2.5)
print(f"instance2.instance_variable: {instance2.instance_variable}")
```

```python
instance2.new_instance_variable_created_outside = 4
print(
    f"instance2.new_instance_variable_created_outside: {instance2.new_instance_variable_created_outside}"
)
```

## Class Variables vs Instance Variables

```python
class MyClass:

    class_variable = "I am a class variable"

    def __init__(self, name) -> None:
        self.instance_variable = name

    def display(self):
        print(f"self.instance_variable = {self.instance_variable}")
        print(f"self.class_variable = {self.class_variable}")
        print(f"MyClass.class_variable = {MyClass.class_variable}")
```

```python
instance1 = MyClass("red")
instance2 = MyClass("blue")
```

```python
instance1.display()
```

```python
instance2.display()
```

```python
id(instance1.class_variable) == id(instance2.class_variable) == id(
    MyClass.class_variable
)
```

## The Mutable Class Variable Trap

```python
class MyClass:

    shared_list = []

    def __init__(self, var) -> None:
        self.shared_list.append(var)

    def display(self):
        print(f"shared_list = {self.shared_list}")
```

```python
instance1 = MyClass("red")
```

```python
instance1.display()
```

```python
instance2 = MyClass("blue")
```

```python
instance2.display()
```

```python
print(id(instance1.shared_list))
print(id(instance2.shared_list))
print(id(MyClass.shared_list))
```

## Encapsulation and Naming Conventions

```python
class MyClass:

    public_class_var = "public"
    _protected_class_var = "protected"
    __private_class_var = "private"

    def __init__(self) -> None:
        self.public_instance_var = "public"
        self._protected_instance_var = "protected"
        self.__private_instance_var = "private"
```

```python
instance = MyClass()
```

```python
print(instance.public_class_var)
print(instance.public_instance_var)
```

```python
print(instance._protected_class_var)
print(instance._protected_instance_var)
```

```python
print(instance._MyClass__private_class_var)
print(instance._MyClass__private_instance_var)
```
