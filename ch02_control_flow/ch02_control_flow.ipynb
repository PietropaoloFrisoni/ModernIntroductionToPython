## Conditional Statements: `if`, `elif`, `else`

```python
temperature = 35
temperature = 25
temperature = 5
```

```python
if temperature > 30:
    print("It is a hot day!")
elif temperature > 20:
    print("It is a nice day!")
else:
    print("It is a cold day!")
```

## Comprehensions

```python
# Traditional loop

squares = []

for i in range(1, 6):
    squares.append(i**2)
print(squares)
```

```python
# List comprehension

squares = [i**2 for i in range(1, 6)]
print(squares)
```

```python
even_squares = [i**2 for i in range(1, 12) if i % 2 == 0]
print(even_squares)
```

```python
# [print(value) for value in [10, 20, 30]]

for value in [10, 20, 30]:
    print(value)
```

```python
# list_of_lists = [[1,2,3], [4,5,6], [7,8,9]]

[[i + j for j in range(3)] for i in range(1, 10, 3)]
```

## `while` Loops


```python
count = 0

while count < 5:
    print(count)
    count += 1  # equivalent to count = count + 1
```

## `break`

```python
count = 0

while True:
    print(count)
    count += 1
    if count == 3:
        break
```

```python
for num in range(1, 50):
    if num % 7 == 0:
        print(f"Found first multiple of 7: {num}")
        break
    print(f"Checked {num}")
```

## `continue`

```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(f"Odd number: {i}")
```

## `pass`

```python
for i in range(5):
    pass  # loop runs but does nothing
```

## `else` Clause on Loops

```python
numbers = [2, 4, 6, 7, 8, 10]

for num in numbers:
    if num == 7:
        print("Found 7!")
        break
else:
    print("7 is not in the list!")
```

## `enumerate`

```python
fruits = ["apple", "banana", "cherry"]

for i in range(len(fruits)):
    print(f"i: {i}, fruits[i]: {fruits[i]}")
```

```python
for i, fruit in enumerate(fruits):
    print(f"i: {i}, fruit: {fruit}")
```

```python
for i, fruit in enumerate(fruits, start=1):
    print(f"i: {i}, fruit: {fruit}")
```

## `zip`

```python
names = ["Alice", "Bob", "Charlie"]
scores = [91, 85, 78]

for name, score in zip(names, scores):
    print(f"{name}: {score}")
```

```python
for i, (name, score) in enumerate(zip(names, scores)):
    print(f"{i}: {name} scored {score}")
```

## Structural Pattern Matching (`match`, `case`)

```python
value = 4

match value:
    case 0:
        print("The number is 0")
    case 1:
        print("The number is 1")
    case _:
        print("The number is something else")
```

```python
command = [1, "hello"]

match command:
    case [0, message]:
        print(f"Command 0 with message: {message}")
    case [1, message]:
        print(f"Command 1 with message: {message}")
    case _:
        print("Unknown command")
```
