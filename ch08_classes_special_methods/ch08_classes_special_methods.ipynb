## `__new__`

```python
class MyClass:

    def __new__(cls, *args):
        print("__new__ has been called")
        print(f"args: {args}")
        print(f"cls: {cls}")
        instance = super().__new__(cls)
        print(f"created instance: {instance}")
        return instance

    def __init__(self, *args) -> None:
        print("__init__ has been called")
        print(f"args: {args}")
```

```python
instance = MyClass(1)
```

```python
MyClass
```

## `__call__`

```python
class MyClass:

    def __init__(self, value):
        print("__init__ has been called")
        self.value = value

    def __call__(self, multiplier):
        print("__call__ has been called")
        print(f"multiplier: {multiplier}")
        return self.value * multiplier
```

```python
instance = MyClass(5)
```

```python
instance(3)
# equivalent to:
# instance.__call__(3)
```

## `__repr__` and `__str__`

```python
class MyClass:

    def __init__(self, value):
        print("__init__ has been called")
        self.value = value

    def __repr__(self):
        print("__repr__ has been called")
        return f"MyClass({self.value})"

    def __str__(self):
        print("__str__ has been called")
        return f"MyClass with self.value = {self.value}"
```

```python
instance = MyClass(5)
```

```python
instance
```

```python
print(instance)
```

## `__eq__`

```python
class MyClass:

    def __init__(self, value):
        print("__init__ has been called")
        self.value = value

    def __eq__(self, other):
        print("__eq__ has been called")
        print(f"other: {other}")
        return isinstance(other, MyClass) and self.value == other.value
```

```python
a = MyClass(5)
b = MyClass(5)
c = MyClass(9)
```

```python
a == b
# equivalent to:
# a.__eq__(b)
```

```python
a is b
```

```python
a == c
```

```python
a == b
```

## `__add__` and `__radd__`

```python
class MyClass:

    def __init__(self, value):
        print("__init__ has been called")
        self.value = value

    def __repr__(self):
        print("__repr__ has been called")
        return f"MyClass({self.value})"

    def __add__(self, other):
        print("__add__ has been called")
        if isinstance(other, MyClass):
            return MyClass(self.value + other.value)
        return NotImplemented

    def __radd__(self, other):
        print("__radd__ has been called")
        return MyClass(other + self.value)
```

```python
a = MyClass(5)
b = MyClass(5)
```

```python
c = a + b
# equivalent to:
# a.__add__(b)
```

```python
c
```

```python
c.value
```

```python
10 + a
```

```python
# int.__add__(10, a)

a.__radd__(10)
```

## Decorating Methods

### `@property`

```python
class Temperature:

    def __init__(self, value):
        self._celsius = value

    @property
    def celsius(self):
        return self._celsius

    # equivalent to:
    # def celsius(self):
    #     return self._celsius
    # celsius = property(celsius)

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:
            raise ValueError("Temperature not valid")
        self._celsius = value

    @celsius.deleter
    def celsius(self):
        print("Deleting the value")
        del self._celsius
```

```python
temperature = Temperature(3)
```

```python
temperature.celsius
```

```python
temperature.celsius = 5
```

```python
temperature.celsius
```

```python
del temperature.celsius
```

### `@staticmethod` and `@classmethod`


```python
class MathTools:

    @staticmethod
    def add(x, y):
        return x + y
```

```python
instance = MathTools()
```

```python
instance.add(4, 6)
```

```python
MathTools.add(4, 6)
```

```python
class Animal:

    species_count = 0

    def __init__(self, name) -> None:
        self.name: str = name
        Animal.species_count += 1

    @classmethod
    def create_pair(cls, name1, name2):
        return cls(name1), cls(name2)
```

```python
a = Animal("dog")
```

```python
a.species_count
```

```python
a = Animal("cat")
```

```python
a.species_count
```

```python
Animal.create_pair("dog", "cat")
```

```python
a.species_count
```

## Decorating Classes: `@dataclass`

```python
from dataclasses import dataclass


@dataclass
class ClassWithData:

    name: str = "default"
    numeric_data_1: float = 0.0
    numeric_data_2: float = 0.0
    data = 0.0


# equivalent to:
# class ClassWithData:

#     name: str = "default"
#     numeric_data_1: float = 0.0
#     numeric_data_2: float = 0.0
#     data = 0.0

# ClassWithData = dataclass(ClassWithData)
```

```python
a = ClassWithData()
```

```python
b = ClassWithData(name="example", numeric_data_1=-1.0, numeric_data_2=2.0)
c = ClassWithData(name="example", numeric_data_1=-1.0, numeric_data_2=2.0)
```
