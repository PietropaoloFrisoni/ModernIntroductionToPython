# Intermezzo 2: Mutability

## Immutable

```python
s = "hello"
print(f"Value of the string: {s}, id: {id(s)}")

s = s.upper()
print(f"Value of the string: {s}, id: {id(s)}")
```

## Mutable

```python
lst = [1, 2, 3]
print(f"Value of the list: {lst}, id: {id(lst)}")

lst.append(4)
print(f"Value of the list: {lst}, id: {id(lst)}")
```

## The Connection Between `id` and `is`

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(f"id(a): {id(a)}")
print(f"id(b): {id(b)}")
print(f"id(c): {id(c)}")

print(f"a is b: {a is b}")
print(f"a is c: {a is c}")
print(f"a == c: {a == c}")
```

```python
b.append(99)
print(f"a after modifying through b: {a}")
```

```python
x = None
print(x is None)
```

## Augmented Assignment and Mutability

```python
x = 10
x += 3  # equivalent to x = x + 3
x -= 2  # equivalent to x = x - 2
x *= 3  # equivalent to x = x * 3
x /= 2  # equivalent to x = x / 2
print(x)
```

```python
x = 10
print(f"Before: id={id(x)}")
x += 3  # equivalent to x = x + 3
print(f"After: id={id(x)}")
```

```python
lst = [1, 2, 3]
# Typo in recording: id(x) should be id(lst)
# (you can still check that the memory address is the same)
print(f"Before: id={id(lst)}")
lst += [3]
print(f"After: id={id(lst)}")
```

## `isinstance` and Type Checking

```python
x = 3.14

print(f"Type of x: {type(x)}")
print(type(x) == float)
isinstance(x, float)
```

```python
x = True

print(f"Type of x: {type(x)}")
print(type(x) == int)
isinstance(x, int)
```

```python
x = 3j

print(f"Type of x: {type(x)}")
isinstance(x, (float, int))
```

## Useful Built-in Functions

### `len`

```python
lst = [2, 11, 1, 4]
print(f"Length: {len(lst)}")
s = "Random string"
print(f"Length: {len(s)}")
```

### `sum`

```python
print(f"Sum: {sum(lst)}")
```

### `min` and `max`

```python
print(f"Max: {max(lst)}")
print(f"Min: {min(lst)}")
```

### `sorted`

```python
print(f"Original: {lst}")
print(f"Sorted: {sorted(lst)}")
print(f"Original: {lst}")
```
